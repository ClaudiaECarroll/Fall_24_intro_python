Conditionals

```python
mass = 2

if mass > 3.0:
    print(mass, 'is large')
```

```python
m = 1

if m > 3.0:
    print(m, 'is large')
else:
    print(m, 'is small')
```

```python
a = 5

if a > 9.0:
    print(a, 'is HUGE')
elif a > 3.0:
    print(a, 'is large')
else:
    print(a, 'is small')
```

```python
# Using conditional to update a variable

velocity = 10.0

if velocity > 20.0:
    print('moving too fast')
else:
    print('adjusting velocity')
    velocity = 50.0
```

```python
print(velocity)
```

```python
# What will this print?

pressure = 71.9

if pressure > 50.0:
    pressure = 25.0
elif pressure <= 50.0:
    pressure = 0.0
print(pressure)
```

Demo 2: For loops

```python
for number in [2, 3, 5]:
    print(number)
```

```python
for letter in "Washington":
  print(letter)
```

```python
store_shelf = ["apples", "bread", "cookies", "avocados", "eggs", "milk", "carrots"]

for item in store_shelf:
    print(item)
```

```python
# The loop variable can be anything

for kitten in [2, 3, 5]:
    print(kitten)
```

```python
# We can create local variables within the for loop

primes = [2, 3, 5]

for p in primes:
    squared = p ** 2
    cubed = p ** 3
    x = [p, squared, cubed]
    print(x[-1])
```

```python
for number in range(0, 3):
    print(number)
```

```python
length = 0

names = ['Curie', 'Darwin', 'Turing']
for value in names:
    length = length + 1

print('There are', length, 'names in the list.')
```

```python
# Accumulating data across loops

total = 0
for char in "tin":
    total = total + 1
    print(total)
```

```python
#Indentation matters!

total = 0
for char in "tin":
    total = total + 1

print(total)
```
